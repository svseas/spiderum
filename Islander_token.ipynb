```python
import pandas as pd
import numpy as np
from mpmath import *
import matplotlib.pyplot as plt
```

```python
mp.dps = 15; mp.pretty = True
def f(param_gamma, param_lambda, j):
    return nsum(lambda n: (1 / (param_gamma + 1 / (1 + n ** param_lambda)) ** n), [0, j])

def L_ratio(gamma_value, lambda_value, j):
  L_inf_value = f(gamma_value, lambda_value, inf)
  L_j_value = f(gamma_value, lambda_value, j)
  L_ratio = L_j_value/L_inf_value
  return L_ratio

def gen_L_ratio(list_1, list_2, j):
  Data = []
  for x in list_1:
    for y in list_2:
      a = list((x, y, L_ratio(x,y,j)))
      Data.append(a)
  return Data




```
